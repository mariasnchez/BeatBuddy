## Emoción: Triste

Sacamos los **metadatos** de las canciones de una playlist.

```python
playlist_id = '37i9dQZF1DXdZjf8WgcTKM'

try:
    playlist_tracks = sp.playlist_items(playlist_id)
except:
    print("No se ha encontrado la playlist")
```

Guardamos en un array valores que nos pueden ser interesantes para listar las canciones como lo son el ID, el nombre(meramente informativo) y la fecha de salida de la canción.

```python
canciones = []

for tracks in playlist_tracks['items']:
    canciones.append([tracks['track']['id'], tracks['track']['name'], tracks['track']['album']['release_date']])

for cancion in canciones:
    print(f'Nombre de la cancion: {cancion[1]}; ID de la cancion: {cancion[0]}; Fecha de lanzamiento: {cancion[2]}')
```

En este caso, vamos a sacar los **datos objetivos** de la canción

```python
parametros_tracks_playlist = []

try:
    for cancion in canciones:
        parametros_tracks_playlist.append(sp.audio_features(cancion[0]))

    for parametros_track in parametros_tracks_playlist:
        print("Parametros de la cancion:", parametros_track)
except:
    print("Ha habido un error")
```

Guardamos los valores en un dataset

```python
import pandas as pd

parametros_tracks_playlist = []

try:
    for cancion in canciones:
        parametros = sp.audio_features(cancion[0])
        if parametros:
            parametros_tracks_playlist.append(parametros[0])
except:
    print("Ha habido un error")

# Crear DataFrame
df = pd.DataFrame(parametros_tracks_playlist)
df['release_date'] = [cancion[2] for cancion in canciones] # Añade la fecha de salida de la canción

# Agregar la columna "mood" con el valor "sad"
df['mood'] = 'sad'

# Visualizar el DataFrame
df
```

```python
media_parametros_playlist = []

media_danceability = 0
media_energy = 0
media_loudness = 0
media_speechiness = 0
media_acousticness = 0
media_instrumentalness = 0
# media_liveness = 0
media_valence = 0
media_tempo = 0

for parametros_track in parametros_tracks_playlist:
    media_danceability += parametros_track["danceability"]
    media_energy += parametros_track["energy"]
    media_loudness += parametros_track["loudness"]
    media_speechiness += parametros_track["speechiness"]
    media_acousticness += parametros_track["acousticness"]
    media_instrumentalness += parametros_track["instrumentalness"]
    media_valence += parametros_track["valence"]
    media_tempo += parametros_track["tempo"]

media_parametros_playlist.append(media_danceability / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_energy / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_loudness / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_speechiness / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_acousticness / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_instrumentalness / len(parametros_tracks_playlist))
# media_parametros_playlist.append(media_liveness / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_valence / len(parametros_tracks_playlist))
media_parametros_playlist.append(media_tempo / len(parametros_tracks_playlist))

print(media_parametros_playlist)
```
